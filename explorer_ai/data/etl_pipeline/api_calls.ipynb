```python
encyclopedia_sources = {
    '1': 'advanced',
    '31': 'intermediate',
    '45': 'concise'
    }
```

```python
# get encyclopedia articles

articles_in_file = '../cache/eb_articles.json'
articles_out_path = '../../knowledgebase/encyclopedia/'

with open(articles_in_file, 'r') as file:
    articles = json.load(file)

for article in articles:
    article_id = article['articleId']
    source_id = str(article['articleTypeId'])
    source = encyclopedia_sources[source_id]
    get_ency.get_article(article_id, source, articles_out_path)
```

```python
files = os.listdir(articles_out_path)
print('Number of articles: {}'.format(len(files)))

filesizes = []
for file in files:
    filepath = os.path.join(articles_out_path, file)
    filesize = os.stat(filepath).st_size
    filesizes.append(filesize)

avg_file_size_kb = round(np.mean(filesizes) / 1000)
print('Average article file size: {} KB'.format(avg_file_size_kb))
```

```python
# flatten primary and related search terms

search_terms_file = '../etl_pipeline/cache/search_terms.json'
with open(search_terms_file, 'r') as f:
    search_terms = json.load(f)

flat_search_terms = []
for item in search_terms:
    these_terms = [item['primary']] + item['related']
    flat_search_terms += these_terms

unique_search_terms = []
[unique_search_terms.append(x) for x in flat_search_terms if x not in unique_search_terms]
pp.pprint(unique_search_terms)
```

```python
# get dictionary entries

dict_out_path = '../../knowledgebase/dictionary/'
data = get_dict.get_dict_data(unique_search_terms, dict_out_path)
```

```python
files = os.listdir(dict_out_path)
print('Number of dictionary entries: {}'.format(len(files)))

filesizes = []
for file in files:
    filepath = os.path.join(dict_out_path, file)
    filesize = os.stat(filepath).st_size
    filesizes.append(filesize)

avg_file_size_kb = round(np.mean(filesizes) / 1000)
print('Average dictionary file size: {} KB'.format(avg_file_size_kb))
```

```python
# get thesaurus entries
# NOTE: this counts responses which give suggested entries -- rerun to remove

thes_out_path = '../knowledgebase/thesaurus/'
data = get_thes.get_thes_data(unique_search_terms, thes_out_path)
```

```python
files = os.listdir(thes_out_path)
print('Number of thesaurus entries: {}'.format(len(files)))

filesizes = []
for file in files:
    filepath = os.path.join(thes_out_path, file)
    filesize = os.stat(filepath).st_size
    filesizes.append(filesize)

avg_file_size_kb = round(np.mean(filesizes) / 1000)
print('Average thesaurus file size: {} KB'.format(avg_file_size_kb))
```
