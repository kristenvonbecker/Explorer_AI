```python
# print the summary stats for the article filesizes (in KB)

path = 'knowledgebase/encyclopedia/'

num_kb = []
for filename in os.listdir(path):
    if filename.startswith('.'):
        continue
    filepath = os.path.join(path, filename)
    filesize = os.stat(filepath).st_size
    filesize_kb = filesize / 1000
    num_kb.append(filesize_kb)

num_bytes_pd = pd.Series(num_kb, name='size_kb')
num_bytes_pd.describe().round(1)
```

```python
# for each article already in the dataset, fetch the article ids for articles referenced
# only fetches from the p tag at the main article level
# does not fetch from any heading sections (or p tags contained in them)

sugg_article_ids = []
existing_article_ids = []

for filename in os.listdir(path):
    if filename.startswith('.'):
        continue

    article_id = filename.split('.')[0]
    existing_article_ids.append(article_id)

    filepath = os.path.join(path, filename)
    with open(filepath, 'r') as infile:
        data = json.load(infile)

    article_p = data['article']['p']

    for p in article_p:
        try:
            xrefs = p['xref']
            if isinstance(xrefs, list):
                for xref in xrefs:
                    sugg_article_ids.append(xref['@articleid'])
            else:
                sugg_article_ids.append(xrefs['@articleid'])

        except (TypeError, KeyError):
            continue

unique_sugg_articles = []
[unique_sugg_articles.append(x) for x in sugg_article_ids if x not in unique_sugg_articles and x not in existing_article_ids]

print('{} new articles\n'.format(len(unique_sugg_articles)))
pp.pprint(unique_sugg_articles)
```

```python
# print random suggested article

rand_article_id = int(random.choice(unique_sugg_articles))

rand_article = get_ency.get_article(rand_article_id, 'advanced')
pp.pprint(rand_article)
```
